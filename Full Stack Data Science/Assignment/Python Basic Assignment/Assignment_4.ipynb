```python
a=[1,2,3,4]
b=a
```

```python
a=[1,2,3,4]
b=a.copy()
```

```python
a=[[1,2,3,4],2,3,]
b=a.copy()
```

```python
a[0][0]=88
b
```

```python
import copy
a=[[1,2,3,4],2,3,]
b=copy.deepcopy(a)
```

```python
a[0][0]=88
b
```
