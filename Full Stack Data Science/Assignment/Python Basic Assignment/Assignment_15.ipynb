```python
7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...
```

```python
def genPrimes():
    primes = []   
    num = 1      
    while True:
        num += 1
        for p in primes:
            if num % p == 0:
                break
        else:
            primes.append(num)
            yield num
```

```python
p=genPrimes()
```
