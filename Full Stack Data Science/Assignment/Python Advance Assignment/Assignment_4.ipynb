```python
Q2. In what contexts do the two operator overloading methods manage printing?

__repr__ and __str__ can be manage printing
```

```python
class Demo:
    def __init__(self):
        pass

    def __repr__(self):
        return 'In __repr__'

Demo()
```

```python
class Demo:
    def __init__(self):
        pass
    def __str__(self):
        return "In __str__"
print(Demo())
```

```python
Q3. In a class, how do you intercept slice operations?
```

```python
class Demo:
    def __init__(self, string):
        self.string = string

    
    def slicing(self, i, j):
        return self.string.__getitem__(slice(i, j,1)) 

d= Demo("feffhuwdud")

sliced = d.slicing(0,2)
  
print(sliced)
```

```python
Q4. In a class, how do you capture in-place addition?
```

```python
class Demo:

    def __init__(self, value):
        self.value = value
    def __iadd__(self, other):
        self.value += other
        
        return self.value
```

```python
d=Demo(5)

d+=2
print(d)
```
