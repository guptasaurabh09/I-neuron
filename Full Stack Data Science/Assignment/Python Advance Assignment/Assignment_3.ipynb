```python
1. What is the concept of an abstract superclass?

calling and absractmethod using a super this is to take the property of abstract method inside a function
```

```python
import abc
class B(object):
    __metaclass__ = abc.ABCMeta
    @abc.abstractmethod
    def foo(self):
        print( 'In B')

class C(B):
    def foo(self):
        super(C, self).foo()
        print ('In C')

C().foo()
```
